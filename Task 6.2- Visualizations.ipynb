## 2. Correlation Matrix Heatmap

```python
# Dropping unneeded columns and creating a subset.
```

```python
df_sub = df_all.drop(['Region', 'Year'], axis = 1)
```

```python
df_sub.head()
```

```python
# Correlation table between variables
df_sub.corr()
```

```python
# Creating subplot using matplotlib
f,ax = plt.subplots(figsize=(10,6))

# Adjusting the font size 
plt.xlabel('X Label', fontsize=2) 
plt.ylabel('Y Label', fontsize=2)

# Creating matrix heatmap using seaborn
corr = sns.heatmap(df_sub.corr(), annot = True, ax = ax)

# Export Heatmap
plt.savefig('heatmap.png') 
```

Observation:

Happiness rank and happiness score have a negative 0.99 correlation. This makes a lot of sense, because as the happiness score increases the happiness rank decreases; given the fact that the highest happiness rank is 1. 

There is also a strong positive correlation between a high happiness score and the following variables: economy (GDP per capita), family, health (life expectancy) and freedom. And a weaker correlation between happiness score and the remaining variables (trust-government corruption and generosity).

## 4. Pair plot

I decided to create two pair plots one with the strong correlation variable and one with the weak correlation variables. 

```python
# Creating smaller subsection to just see the strongest variables
df_sub_strong = df_all.drop(['Region', 'Year','Generosity', 'Trust (Government Corruption)', 'Freedom'], axis = 1)
```

```python
pair_plot_strong = sns.pairplot(df_sub_strong)
```

```python
# Creating smaller subsection to just see the weakest variables
df_sub_weak = df_all.drop(['Region', 'Year','Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)'], axis = 1)
```

After creating the subset for the weak correlation variables I received an error while trying to create the pair plot that the sub set had duplicates.

```python
# Check for duplicate indices
duplicates = df_sub_weak.index.duplicated()
print(df_sub_weak[duplicates])
```

```python
# Drop duplicates
df_sub_weak = df_sub_weak[~duplicates]
```

```python
# Recreate pairplot
pair_plot_weak = sns.pairplot(df_sub_weak)
```

observation:

The strong pair plot looked exactly as expected. The plots shows strong correlations between all variables. 

## 5. Categorical plot

```python
# Creating a histogram to see distribution
sns.histplot(df_all['Happiness Score'], bins = 10, kde = True)
```

```python
# Creating a score category for categorical plot
df_all.loc[df_all['Happiness Score'] < 4.5 , 'Score category'] = 'Low score'
```

```python
df_all.loc[(df_all['Happiness Score'] >= 4.5) & (df_all['Happiness Score'] < 6.5), 'Score category'] = 'Middle score'
```

```python
df_all.loc[df_all['Happiness Score'] >= 6.5, 'Score category'] = 'High score'
```

```python
df_all['Score category'].value_counts(dropna = False)
```

```python
# Categorical plot with Economy (GDP per Capita)
sns.set(style="ticks")
cat_plot_GDP = sns.catplot(x='Economy (GDP per Capita)', y='Happiness Score', hue="Score category", data=df_all)
```

```python
# Creating categorical plot with family
sns.set(style="ticks")
cat_plot_family = sns.catplot(x='Family', y='Happiness Score', hue="Score category", data=df_all)
```

```python
# Creating categorical plot with Health
sns.set(style="ticks")
cat_plot_health = sns.catplot(x='Health (Life Expectancy)', y='Happiness Score', hue="Score category", data=df_all)
```

```python
# Creating categorical plot with freedom
sns.set(style="ticks")
cat_plot_freedom = sns.catplot(x='Freedom', y='Happiness Score', hue="Score category", data=df_all)
```

```python
# Creating categorical plot with generosity
sns.set(style="ticks")
cat_plot_generosity = sns.catplot(x='Generosity', y='Happiness Score', hue="Score category", data=df_all)
```

Observation:

These several categorical plots don't show a significant line between low, middle and high happiness score. There also don't seem to be any outliers across the variables.

On the other hand, freedom, generosity and trust are the variables that contribute the least to the happiness score.
